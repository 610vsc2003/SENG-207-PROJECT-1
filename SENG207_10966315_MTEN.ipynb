# Replacing null values in each column.

```python

eachColumn = ['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol','quality']
for column in eachColumn:
    data[column].fillna(data[column].mean(), inplace=True)
```

# METHOD: Replacing null values with the mean of each column.
# Reason for replacing null values using the mean of each column:
Replacing null values with the mean of the column is a common technique used in data analysis when missing data is present. Deleting rows with missing values can lead to loss of information and reduce the sample size, which may negatively impact the accuracy and representativeness of the analysis.

By replacing null values with the mean of the column, we can preserve the sample size and provide a reasonable estimate of the missing data. This approach assumes that the data is normally distributed and that the missing values are missing at random, meaning that the probability of a value being missing is unrelated to its true value.

However, this approach may not be appropriate in all cases. For example, if the missing values are not missing at random or if the data is not normally distributed, replacing null values with the mean of the column may not be an accurate representation of the missing data. In such cases, alternative methods such as interpolation, imputation or dropping rows/columns may be more appropriate.

# Line graph indicating the quality of column in dataset.

```python

eachColumn = ['fixed acidity', 'volatile acidity', 'citric acid',
       'residual sugar', 'chlorides', 'free sulfur dioxide',
       'total sulfur dioxide', 'density', 'pH', 'sulphates', 'alcohol']

for column in eachColumn:
    plt.title(f'Quality graph for {column}')
    sns.lineplot(x='quality', y= column, data = data)
    plt.figure(figsize=(12,7))

    plt.show()
```

# Depicting the correllation between other characteristics and quality

```python
acids = ['fixed acidity', 'volatile acidity', 'citric acid']
for acid in acids:
    plt.title(f'Correlation between the pH and {acid}')
    sns.regplot(x=acid ,y='pH' , data=data, line_kws={'color':'green'})
    plt.figure(figsize=(12,10))
    plt.show()
```
